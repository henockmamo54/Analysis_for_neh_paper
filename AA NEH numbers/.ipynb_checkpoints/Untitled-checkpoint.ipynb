```python
import os
import numpy as np
import pandas as pd
from scipy.optimize import nnls
from matplotlib import pyplot as plt
from sklearn.metrics import r2_score
import math
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde
```

```python
data=pd.read_csv("C:\\Workplace\\C++\\d2ome_v2\\v2\\v2\\bin\\Debug\\Estimated_AA_neh_numbers.csv")
data.columns
```

```python
cols=['Asym_NEH', 'APE', 'MPE']
col_cis=['Asym_NEH_ci', 'APE_ci','MPE_ci']

for i in range(3):
    col=cols[i]
    col_ci=col_cis[i] 
    print("============> ",col,"\t")
    plt.rc('font', size=15) 
    plt.rc('figure', figsize=(8,5)) 
    plt.errorbar(range(20), data[col], data[col_ci],fmt ='o',mfc='k',color="r",ms=6)
    plt.xticks(range(20),[x.upper() for x in ['a', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'k', 'l', 'm', 'n', 'p','q', 'r', 's', 't', 'v', 'w', 'y']])
    plt.ylabel("NEH")    
    plt.savefig(col+"_CI.jpeg",dpi=900,bbox_inches='tight')
    plt.show()
```
