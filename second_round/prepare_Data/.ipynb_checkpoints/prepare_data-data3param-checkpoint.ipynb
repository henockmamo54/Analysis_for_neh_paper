```python
sum_i0=(res['I0'] + res['I1']+res['I2']+res['I3']+res['I4']+res['I5'])
sum_i0_1=(res['I0.1'] + res['I1.1']+res['I2.1']+res['I3.1']+res['I4.1']+res['I5.1'])
sum_i0_2=(res['I0.2'] + res['I1.2']+res['I2.2']+res['I3.2']+res['I4.2']+res['I5.2'])
sum_i0_3=(res['I0.3'] + res['I1.3']+res['I2.3']+res['I3.3']+res['I4.3']+res['I5.3'])
sum_i0_4=(res['I0.4'] + res['I1.4']+res['I2.4']+res['I3.4']+res['I4.4']+res['I5.4'])

sum_i0_11=(res['I0.11'] + res['I1.11']+res['I2.11']+res['I3.11']+res['I4.11']+res['I5.11'])
```

```python
res['I0']=res['I0'].astype('float')/sum_i0
res['I1']=res['I1'].astype('float')/sum_i0
res['I2']=res['I2'].astype('float')/sum_i0
res['I3']=res['I3'].astype('float')/sum_i0
res['I4']=res['I4'].astype('float')/sum_i0
res['I5']=res['I5'].astype('float')/sum_i0

res['I0.11']=res['I0.11'].astype('float')/sum_i0_11
res['I1.11']=res['I1.11'].astype('float')/sum_i0_11
res['I2.11']=res['I2.11'].astype('float')/sum_i0_11
res['I3.11']=res['I3.11'].astype('float')/sum_i0_11
res['I4.11']=res['I4.11'].astype('float')/sum_i0_11
res['I5.11']=res['I5.11'].astype('float')/sum_i0_11
```

```python
def updatedRIAs(data):
    for i in range(1,11):
#         data=data[[isinstance(x,float) for x in data[f'I0.{i}']]] 
        data=data[[ not(type(x)==str) for x in data[f'I{0}.'+str(i)] ]]
    
    for i in range(1,11):
        sum_i0=0
        for j in range(0,6): 
            sum_i0=data[f'I{j}.'+str(i)].astype('float')

        for j in range(0,6):
            data[f'I{j}.'+str(i)]=data[f'I{j}.'+str(i)].astype('float')/sum_i0
    return data
res=updatedRIAs(res)
```
