```python
def plot_spec(data,time,color): 

    plt.figure(figsize=(5,4))
    plt.rc('font', size=14) 


    ax=plt.subplot(1,1,1)
    temp=data
    temp_sum=sum(temp) 

    xvals=[ f'{575.29 + x/3:.1f}' for x in range(6)] 
#     (markerline, stemlines, baseline) = plt.stem(temp,linefmt ='k',markerfmt='ko')
    (markerline, stemlines, baseline) = plt.stem(temp,linefmt =color,markerfmt=color+'o')

    plt.setp(baseline, visible=False,alpha=0.5)
    ax.set_xticks([0,1,2,3,4,5])
    ax.set_xticklabels(xvals) 
    plt.xlabel("m/z") 
#     plt.ylabel("RIA")
    plt.text(0,0.03,r"M$_0$")
    plt.text(1.01,0.03,r"M$_1$")
    plt.text(2.01,0.03,r"M$_2$")
    plt.text(3.01,0.03,r"M$_3$")
    plt.text(4.01,0.03,r"M$_4$")
    plt.text(5.01,0.03,r"M$_5$") 
    plt.box()
    plt.savefig(f"spec_plot_{40}.jpeg",dpi=900,bbox_inches = "tight")
    plt.show()
```

```python
def plot_spec(data,time,color,offset): 

    plt.figure(figsize=(6,4))
    plt.rc('font', size=16) 


    ax=plt.subplot(1,1,1)

    xvals=[ f'{offset + 575.29 + x/2:.1f}' for x in range(6)] 
#     (markerline, stemlines, baseline) = plt.stem(temp,linefmt ='k',markerfmt='ko')

    print(xvals)
    if offset == 0.5:
        xvals=[ f'{x:.1f}' for x in [575.29]]+xvals
        data=[-10]+data
        print(xvals,data)
        
    
    temp=data
    temp_sum=sum(temp) 
    (markerline, stemlines, baseline) = plt.stem(temp,linefmt =color,markerfmt=color+'o')

    plt.setp(baseline, visible=False,alpha=0.5)
    if offset == 0:
        ax.set_xticks([0,1,2,3,4,5])
    else:        
        ax.set_xticks([0,1,2,3,4,5,6])
    ax.set_xticklabels(xvals) 
    plt.xlabel("m/z") 
#     plt.ylabel("RIA")
    label_ofset=1
    plt.text(label_ofset+0,0.03,r"M$_0$")
    plt.text(label_ofset+1.01,0.03,r"M$_1$")
    plt.text(label_ofset+2.01,0.03,r"M$_2$")
    plt.text(label_ofset+3.01,0.03,r"M$_3$")
    plt.text(label_ofset+4.01,0.03,r"M$_4$")
    plt.text(label_ofset+5.01,0.03,r"M$_5$") 
    plt.savefig(f"spec_plot_{time}.jpeg",dpi=900,bbox_inches = "tight")
    plt.xlim([-0.5,6.2])
    plt.ylim([0,0.5])    
    plt.box()
    plt.savefig('2.jpeg',dpi=900, bbox_inches = "tight")
    plt.show()

# dummy
data=[0.4841,0.3342,0.1324,0.0383,0.0089,0.0018]
# plot_spec(data,"exp",'g',0)
plot_spec(data,"exp",'m',1/2)
# plot_spec(data,"exp",'r',1)
```

```python
def plot_spec(data,time,color,offset): 

    plt.figure(figsize=(7,4))
    plt.rc('font', size=16) 


    ax=plt.subplot(1,1,1)

    xvals=[ f'{offset + 575.29 + x/2:.1f}' for x in range(6)] 
#     (markerline, stemlines, baseline) = plt.stem(temp,linefmt ='k',markerfmt='ko')

    print(xvals)
    
    xvals=[ f'{x:.1f}' for x in [575.29,575.29+0.5]] +xvals 
    data=[-10]*2+data
    print(xvals,data)
        
    
    temp=data
    temp_sum=sum(temp) 
    (markerline, stemlines, baseline) = plt.stem(temp,linefmt =color,markerfmt=color+'o')
    print(xvals)
    plt.setp(baseline, visible=False,alpha=0.5)
            
    ax.set_xticks([0,1,2,3,4,5,6,7])
    ax.set_xticklabels(xvals) 
    plt.xlabel("m/z") 
#     plt.ylabel("RIA")
    label_ofset=2
    plt.text(label_ofset+0,0.03,r"M$_0$")
    plt.text(label_ofset+1.01,0.03,r"M$_1$")
    plt.text(label_ofset+2.01,0.03,r"M$_2$")
    plt.text(label_ofset+3.01,0.03,r"M$_3$")
    plt.text(label_ofset+4.01,0.03,r"M$_4$")
    plt.text(label_ofset+5.01,0.03,r"M$_5$") 
    plt.savefig(f"spec_plot_{time}.jpeg",dpi=900,bbox_inches = "tight")
    plt.xlim([-0.5,7.2])
    plt.ylim([0,0.5])    
    plt.box()
    plt.savefig('3.jpeg',dpi=900, bbox_inches = "tight")
    plt.show()

# dummy
data=[0.4841,0.3342,0.1324,0.0383,0.0089,0.0018]
# plot_spec(data,"exp",'g',0)
# plot_spec(data,"exp",'y',1/2)
plot_spec(data,"exp",'m',1)
```
