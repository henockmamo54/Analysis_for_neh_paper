```python
def plot_spec(data,time,color): 

    plt.figure(figsize=(8,6))
    plt.rc('font', size=14) 


    ax=plt.subplot(1,1,1)
    temp=data
    temp_sum=sum(temp) 

    xvals=[ f'{575.29 + x/3:.2f}' for x in range(6)] 
#     (markerline, stemlines, baseline) = plt.stem(temp,linefmt ='k',markerfmt='ko')
    (markerline, stemlines, baseline) = plt.stem(temp,linefmt =color,markerfmt=color+'o')

    plt.setp(baseline, visible=False)
    ax.set_xticks([0,1,2,3,4,5])
    ax.set_xticklabels(xvals) 
    plt.xlabel("m/z") 
    plt.ylabel("RIA")
    plt.text(0,0.03,r"M$_0$")
    plt.text(1.01,0.03,r"M$_1$")
    plt.text(2.01,0.03,r"M$_2$")
    plt.text(3.01,0.03,r"M$_3$")
    plt.text(4.01,0.03,r"M$_4$")
    plt.text(5.01,0.03,r"M$_5$") 
    plt.savefig(f"spec_plot_{time}.jpeg",dpi=900,bbox_inches = "tight")
    plt.show()
```
