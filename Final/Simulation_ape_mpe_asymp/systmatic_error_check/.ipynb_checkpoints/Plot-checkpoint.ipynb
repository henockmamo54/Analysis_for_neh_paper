```python
import os
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
# data_with_shift.columns
```

```python
aa_sh_name={'Q':'Gln (Q)','N':'Asn (N)','S':'Ser (S)','G':'Gly (G)','T':'Thr (T)','A':'Ala (A)','D':'Asp (D)','P':'Pro (P)',
            'M':'Met (M)','V':'Val (V)','E':'Glu (E)','H':'His (H)','K':'Lys (K)','L':'Leu (L)','I':'Ile (I)','F':'Phe (F)',
            'C':'Cys (C)','Y':'Tyr (Y)','R':'Arg (R)','W':'Trp (W)',}
```

```python
data_with_shift=pd.read_csv('res_with_shift.csv')
data_with_shift=data_with_shift.sort_values(by='neh') 

data_with_no_shift=pd.read_csv('res_with_no_shift.csv')
data_with_no_shift=data_with_no_shift.sort_values(by='neh') 

data_res_with_no_noise=pd.read_csv('res_with_no_noise.csv')
data_res_with_no_noise=data_res_with_no_noise.sort_values(by='neh') 
```

```python
def plot(data,filename):
    temp_pd=pd.DataFrame()
    for i in range(20):
        new_pd=pd.DataFrame()
        new_pd['val']=list(data.iloc[i,2:])
        new_pd['AA']=data.iloc[i,0]
        temp_pd=pd.concat([temp_pd,new_pd])

    PROPS = {
        'boxprops':{'facecolor':'none', 'edgecolor':'black'},
        'medianprops':{'color':'black'},
        'whiskerprops':{'color':'black'},
        'capprops':{'color':'black'}
    }

    plt.rc('font', size=14) 
    plt.rc('figure', figsize=(12,8)) 
    sns.boxplot(data=temp_pd,x='AA',y='val',**PROPS,saturation=1, width=0.5)
    plt.scatter(data.aa,data.neh,c='r',zorder=10)
    plt.ylabel(r'$N_{EH}$')
    plt.xlabel('')
    plt.xticks(range(20),[aa_sh_name[x] for x in data.aa],rotation=90)
    plt.savefig(filename+".jpeg",dpi=900,bbox_inches='tight')
    plt.show()
```

```python
plot(data_res_with_no_noise,"res_with_no_noise")
```

```python
plot(data_with_shift,"data_with_shift")
```

```python
plot(data_with_no_shift,"data_with_no_shift")
```

```python
data=pd.read_csv('data.csv') 

plt.rc('font', size=14) 
plt.figure(figsize=(10,8)) 
res=dict()

for pep in list(data['PeptidesPassedNEHfilters.Peptide']):
    visited=set()
    for c in (pep):
        if c not in visited:
            res[c.upper()]=res.get(c.upper(),0)+1
            visited.add(c)
res=np.array(sorted(res.items(),key= lambda item:item[1],reverse=False)) 

temp=range(0,20,1)
plt.ylabel("Count")
plt.bar([x for x in temp],[int(x) for x in (res[:,1])],alpha=0.85,color='k')

# plt.xticks([x for x in temp],res[:,0])
plt.xticks(range(20),[aa_sh_name[x] for x in res[:,0]],rotation=90)

plt.savefig("count.jpeg",dpi=900,bbox_inches='tight')
plt.show()
```
