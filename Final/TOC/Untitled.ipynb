```python
import os 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
os.listdir()
```

```python
exp=pd.read_csv('EXP_DATA.csv')
theo=pd.read_csv('THEO_DATA.csv')
```

```python
plt.rc('font', size=15) 
plt.rc('figure', figsize=(15,9)) 



plt.plot(theo['DataPoint._xValue'],
         [ float((x).split('=')[-1].replace('}','').replace('{','').split(',')[0]) for x in theo['DataPoint']],
        lw=3,c='#F57C00',ls=':')

plt.scatter(exp['DataPoint._xValue'],
            [ float((x).split('=')[-1].replace('}','').replace('{','').split(',')[0]) for x in exp['DataPoint']]
            ,marker='o',s=90,c='#3F51B5')

plt.vlines(x=-0.5,ymin=0.11,ymax=0.51,color='k',alpha=0.5)
plt.hlines(y=0.11,xmin=-0.5,xmax=32,color='k',alpha=0.5)

plt.box()
plt.xlabel('Labeling duration (Days)')
plt.ylabel('Relative abundance of monoistope (RIA)')

plt.savefig("Sample_plot.jpeg",dpi=900,bbox_inches='tight')
plt.show()
```
