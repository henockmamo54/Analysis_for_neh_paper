```python
import os 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
files_info=[(x, float(x.split('_')[-1].replace('.csv',''))) for x in os.listdir() if '_AA_neh_numbers' in x]
```

```python
ala=[]
leu=[]
gly=[]
glu=[]
for file,time in files_info:
    print(file)
    data=pd.read_csv(file)
    ala.append(float(data[data.AA_name=='Alanine (A)'].Asym_NEH))
    leu.append(float(data[data.AA_name=='Leucine (L)'].Asym_NEH)) 
    gly.append(float(data[data.AA_name=='Glycine (G)'].Asym_NEH))
    glu.append(float(data[data.AA_name=='Glutamic acid (E)'].Asym_NEH))
    
```

```python
last_data=pd.read_csv(r"C:\Workplace\Python\AnalysisForThePaper\NEH\Final\Experimental_result\liverpool_liver\Estimated_AA_neh_numbers_liverpool_liver_.csv")
ala.append(float(last_data[last_data.AA_name=='Alanine (A)'].Asym_NEH))
leu.append(float(last_data[last_data.AA_name=='Leucine (L)'].Asym_NEH)) 
gly.append(float(last_data[last_data.AA_name=='Glycine (G)'].Asym_NEH))
glu.append(float(last_data[last_data.AA_name=='Glutamic acid (E)'].Asym_NEH))
```

```python
time=[x for f,x in files_info]+[31]
```

```python
plt.figure(figsize=(7,6))
plt.rc('font', size=12)

plt.scatter(time,ala,label='Alanine (A)')
plt.scatter(time,leu,label='Leucine (L)')
plt.scatter(time,gly,label='Glycine (G)')
plt.scatter(time,glu,label='Glutamic acid (E)')
plt.xticks(time)
plt.legend(loc='best', bbox_to_anchor=(0.65, 0.0, 0.5, 1))
plt.xlabel("Labeling duration (Days)")
plt.ylabel('$N_{EH}$')
plt.savefig("aa.jpeg",dpi=900)
```
