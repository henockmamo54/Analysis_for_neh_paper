```python
import os
import math
import random
import warnings
import numpy as np
import pandas as pd 
from sklearn import metrics
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats as st
warnings.filterwarnings("ignore")
```

```python
NEH_LIVERPOOL={'A' : 3.442,'C' : 1.811,'D' : 2.974,'E' : 4.378,'F' : 0.519,'G' : 1.822,'H' : 1.957,'I' : 0.76,'K' : 0.037,
               'L' : 0.92,'M' : 0.774,'N' : 1.319,'P' : 1.673,'Q' : 3.985,'R' : 1.701,'S' : 1.972,'T' : 0.351,'V' : 0.742,
               'W' : 0.277,'Y' : 0.957}
def get_newNeh(pep):
    res=0
    for c in pep:
        res+=NEH_LIVERPOOL[c.upper()]
    return res
 
def get_aa_count(peps):
    res={}
    
    for pep in peps:
        for c in np.unique(list(pep)):
            res[c.upper()]=  res.get(c.upper(),0)+1
    df=pd.DataFrame([res.keys(),res.values()]).transpose()
    df.columns=['AA','Count']
    df=df.sort_values(by='Count')
    return df

def get_hspr_d_count(peps):
    res=[[],[]]
    scale=20
    for x in peps:
        hsp_count= scale * (x.count('S')+x.count('P')+x.count('H')+x.count('R'))
        d_count= scale *x.count('D')
        
        if hsp_count > d_count:
            res[0].append(hsp_count)
            res[1].append('g')
        else:
            res[0].append(d_count)
            res[1].append('r')
            
    return res   
    
def q_filter(df,col):
    q_low = df[col].quantile(0.05)
    q_hi  = df[col].quantile(0.95)

    df_filtered = df[(df[col] < q_hi) & (df[col] > q_low)]
    return df_filtered
```

```python
prots=['CPSM_MOUSE', 'AL1L1_MOUSE', 'FAS_MOUSE', 'SARDH_MOUSE', 'PYC_MOUSE', 'DHE3_MOUSE', 'THIM_MOUSE', 'HYES_MOUSE',
       'ATPB_MOUSE', 'HMCS2_MOUSE', 'ASSY_MOUSE', 'ECHA_MOUSE', 'ATPA_MOUSE', 'CH60_MOUSE', 'FAAA_MOUSE', 'M2GD_MOUSE',
       'PYGL_MOUSE', 'DHSO_MOUSE', 'EF2_MOUSE', 'CAH3_MOUSE', 'AL1A1_MOUSE', 'BHMT1_MOUSE', 'ADH1_MOUSE', 'SBP1_MOUSE',
       'SAHH_MOUSE', 'ALDH2_MOUSE', 'PGM1_MOUSE', 'EST1D_MOUSE', 'CLH1_MOUSE', 'ARGI1_MOUSE']

res_all=pd.DataFrame()
prot_k_orginal=[]
prot_k_modified=[]
prot_sd_orginal=[]
prot_sd_modified=[]
for prot_name in prots:
    
    data_original=pd.read_csv(r'C:\Workplace\Python\AnalysisForThePaper\NEH\d2ome_output\liverpool_liver\\'+f'{prot_name}.RateConst.csv')
    data_original=data_original.dropna()
    data_original['Protein']=prot_name

    data_modified=pd.read_csv(r'C:\Workplace\executables_test\d2ome_4_1_2024\test_output\test\liverpool_liver_modified\\'+f'{prot_name}.RateConst.csv')
    data_modified=data_modified.dropna()
    data_modified['Protein']=prot_name

    data_original.columns=[x.strip() for x in data_original.columns]
    data_modified.columns=[x.strip() for x in data_modified.columns]    
    
    
#     normal=[x for x in data_original.Peptides if  ((x.count('S')+x.count('P')+x.count('H')+x.count('R')) <=1) and 'D' not in x]
#     under_estimate=[x for x in data_original.Peptides if  ((x.count('S')+x.count('P')+x.count('H')+x.count('R')) >=1) and 'D' not in x  ]
    
    under_estimate =list(data_original.Peptides)
    

    merged=pd.merge(data_original[['Protein','Peptides', 'RateConstants', 'Charge', 'NEH','Rsquared','NDP','sigma']]
                    ,data_modified[['Protein','Peptides', 'RateConstants', 'Charge', 'NEH','Rsquared','sigma']]
                    ,on=['Protein','Peptides','Charge'],
                   suffixes=['_orginal','_modi'])
    merged['NEH_modi']=[get_newNeh(x) for x in merged.Peptides]
    
    
    merged=merged[merged.Rsquared_orginal>0.95]
    merged=merged[merged.NDP>4]    
#     merged=merged[merged.NEH_modi < merged.NEH_orginal]
    

    merged = q_filter(merged,"RateConstants_orginal")
    merged = q_filter(merged,"RateConstants_modi")

    res_all=pd.concat([res_all,merged])
#     res=get_hspr_d_count(merged.Peptides)
    

    try:
        plt.figure(figsize=(16,5))
        plt.subplot(1,2,1)

        x=merged.query(f'Peptides in {under_estimate}').RateConstants_orginal
        y=merged.query(f'Peptides in {under_estimate}').RateConstants_modi    
        print(prot_name)    
        print(f'original k - mode {st.mode(x)[0]}, medain {np.median(x)}, mean {np.mean(x)}. SD {np.std(x)} range [{-np.min(x)+np.max(x)}]')
        print(f'modified k - mode {st.mode(y)[0]}, medain {np.median(y)}, mean {np.mean(y)}. SD {np.std(y)} range [{-np.min(y)+np.max(y)}]')
        
        prot_sd_orginal.append(np.std(x))
        prot_sd_modified.append(np.std(y))
        prot_k_orginal.append(np.median(x))
        prot_k_modified.append(np.median(y))
        
        if np.std(x) > np.std(y):
            print("=============> ***** +++++",  100*abs(np.std(x) - np.std(y))/np.std(x)  )
            
        plt.scatter(x,y,c='k',alpha=0.5)
        plt.plot(x,x,ls=':',c='r')
        plt.xlabel('k (original)')
        plt.ylabel('k (modified)')


#         plt.subplot(1,2,2)
#     #     sns.kdeplot(x,color='green',label=f'original n ={x.shape[0]}',bw=0.75)
#     #     sns.kdeplot(y,color='blue',label=f'modified n ={y.shape[0]}',bw=0.75)
#         sns.histplot(x,color='green',label=f'original n ={x.shape[0]}',alpha=0.15,kde=True,bins=x.shape[0]//5)
#         sns.histplot(y,color='blue',label=f'modified n ={y.shape[0]}',alpha=0.15,kde=True,bins=x.shape[0]//5)
#         plt.legend()
#         plt.xlabel('k')
        plt.show()

#         #===================================================
#         plt.figure(figsize=(16,5))
#         plt.subplot(1,2,1)

#         plt.scatter(merged.NEH_orginal,merged.NEH_modi,alpha=0.5,s=res[0],c=res[1])    
#         plt.plot(merged.NEH_orginal,merged.NEH_orginal,ls=':',c='m')
#         plt.xlabel('NEH (original)')
#         plt.ylabel('NEH (modified)')

#         plt.subplot(1,2,2)
#         res=get_aa_count(merged.Peptides)
#         plt.bar(range(res.shape[0]),res.Count,color='k')
#         plt.xticks(range(res.shape[0]),res.AA)

#         plt.show()
        
    except Exception as e:
        print(e) 
        continue
    
    print("==========================================================")
#     break
```
